```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/2019.csv")

print("cols: {}\nsize: {}x{}".format(list(df.columns), len(df.columns), df.size))
df.head()
```

```python
#TODO: column documentation on original data
#TODO: select which columns should be dropped and which should be used for analysis
df.drop(['Rk', '2PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'PosRank', 'OvRank', 'PPR', 'Fmb', 'GS'], axis=1, inplace=True)
```

```python
#fix name formatting.
#Used lambda because its the only time it will be used. 
#Used 2 splits because '*' AND '\' are not always present in the string
df['Player'] = df['Player'].apply(lambda x: x.split('*')[0]).apply(lambda x: x.split('\\')[0])
#test for above lambda
print("Pass: {}".format(not bool([s for s in df['Player'] if ('*' in s) or ('\\' in s)])))
```
